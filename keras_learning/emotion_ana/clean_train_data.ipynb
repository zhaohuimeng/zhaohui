```python
# !/usr/bin/env python
# -*- coding: utf-8 -*-

## abandon ##

import sys

def word_filter(line_data):
    if len(line_data) == 0:
        return line_data
    
    ind = 1
    spstr = "回复"
    w_cnt = 0
    while(ind >= 0 and w_cnt <= 10):
        w_cnt += 1
        ind = line_data.find('@')
        if ind == -1:
            break
        ind21 = line_data[ind:].find(':')
        ind22 = line_data[ind:].find(' ')
        if ind21 == -1:
            if ind22 == -1:
                ind2 = len(line_data)
            else:
                ind2 = ind22 + ind
        else:
            if ind22 == -1:
                ind2 = ind21 + ind
            else:
                ind2 = min(ind21, ind22) + ind
        ind3 = max(ind-2, 0)
        if spstr.decode('utf-8') == line_data[ind3:ind]:
            line_data = line_data[0:ind3] + line_data[ind2+1:]
        else:
            line_data = line_data[0:ind] + line_data[ind2+1:]

    return line_data


def main():
    
    cnt = 0
    fout1 = open("data_train_clean", "w")
    
    with open("data_train.csv") as f1:
        for line in f1:
            line = line.strip('\n').split('\t')
            
            cnt += 1
            if len(line) != 3:
                #sys.stderr.write("Illegal line: %d\n" % cnt)
                continue
            
            try:
                line_data = line[1].decode('utf-8')
            except:
                #sys.stderr.write("Decode Error: %d\n" % cnt)
                continue
                
            line_data = word_filter(line_data)
            line_data = line_data.strip('')
            if len(line_data) == 0:
                #sys.stderr.write("Empty data: %d\n" % cnt)
                continue
            try:
                line_data = line_data.encode('utf-8')
            except:
                #sys.stderr.write("Encode Error: %d\n" % cnt)
                continue
            
            print >> fout1, "%s\t%s" % (line_data, line[2])
            
    fout1.close()
    print "Successful"
    
    
if __name__ == "__main__":
    main()
            
            
```
