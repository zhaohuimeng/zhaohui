```python
import sys

def load_word_dict(filename):
    word_dict = {}
    with open(filename) as f1:
        for line in f1:
            line = line.strip('\n').split('\t')
            if len(line) != 2:
                sys.stderr.write("dict line error")
                continue
            word = line[1].decode('utf-8')  
            word_dict[word] = line[0]
    return word_dict


def word_clean(line_data):
    if len(line_data) == 0:
        return line_data
    
    ind = 1
    spstr = "回复"
    w_cnt = 0
    while(ind >= 0 and w_cnt <= 10):
        w_cnt += 1
        ind = line_data.find('@')
        if ind == -1:
            break
        ind21 = line_data[ind:].find(':')
        ind22 = line_data[ind:].find(' ')
        if ind21 == -1:
            if ind22 == -1:
                ind2 = len(line_data)
            else:
                ind2 = ind22 + ind
        else:
            if ind22 == -1:
                ind2 = ind21 + ind
            else:
                ind2 = min(ind21, ind22) + ind
        ind3 = max(ind-2, 0)
        if spstr.decode('utf-8') == line_data[ind3:ind]:
            line_data = line_data[0:ind3] + line_data[ind2+1:]
        else:
            line_data = line_data[0:ind] + line_data[ind2+1:]

    return line_data


def parse_data(filename, word_dict, flag):
    MAX_NUMBER = 4060
    pro_ind = 0
    if flag == 0:
        fout = open("train_data_encode", "w")
        line_data_len = 3
    elif flag == 1:
        fout = open("test_data_encode", "w")
        line_data_len = 2
    else:
        sys.stderr.write("flag error")
        return
    
    with open(filename) as f1:
        for line in f1:
            empty_line = 0
            line = line.strip('\n').split('\t')
            
            if len(line) != line_data_len:
                sys.stderr.write("input data format error")
                continue
            try:
                x = line[1].decode('utf-8')
            except:
                sys.stderr.write("input data decode error")
                
            x = word_clean(x)
            x = x.strip()
            if len(x) == 0:
                empty_line = 1
            
            if flag == 0:
                y = line[2]
                try:
                    y = int(y)
                    if y <= -1 or y >= 3:
                        sys.stderr.write("train data y value error")
                        continue
                except:
                    sys.stderr.write("train data y error")
                    continue
            
            pro_ind += 1
            if pro_ind % 500 == 0:
                print "has proc %d: %d" % (flag, pro_ind)
            
            if empty_line == 1:
                if flag == 0:
                    print >> fout, "[0]\t%d" % y
                elif flag == 1:
                    print >> fout, "[0]"
            else:
                outstr = ""
                for i in x:
                    ind = word_dict.get(i, MAX_NUMBER)
                    outstr = outstr + str(ind) + ","
                outstr = outstr.strip(',')
                if flag == 0:
                    print >> fout, "[%s]\t%d" % (outstr, y)
                elif flag == 1:
                    print >> fout, "[%s]" % outstr
    
    fout.close()
            

#def main():
print "Start!"

word_dict = load_word_dict("word_dict")
print "load word dict finish!"

parse_data("data_train.csv", word_dict, 0)
print "parse train data finish!" 

parse_data("data_test.csv", word_dict, 1)
print "parse test data finish!" 
```
